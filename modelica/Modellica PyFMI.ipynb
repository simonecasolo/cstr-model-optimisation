```python
from pyfmi import load_fmu

model = load_fmu("andris_tut.cstr_model.fmu")
opts = model.simulate_options()

opts
```

```python
Ci, Ti, Tci, Qc = [0.97, 351.5, 351.6, 150]

start_values={'UA':7E5,'k_0':7.2E10,'Ci':Ci,'Ti':Ti,'Tci':Tci,'Qc':Qc}

for item in start_values:
    model.set(item, start_values[item])
```

```python
res = model.simulate(start_time=0, final_time=20)
```

```python
print([res['C'][-1], res['T'][-1], res['Tc'][-1]])
```
